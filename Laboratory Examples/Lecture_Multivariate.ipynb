```python
# import the libraries: numpy, pandas, matplotlib, seaborn
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Let us use the mpg of dataset of seaborn
df = sns.load_dataset('mpg')
df
```

```python
# Describe the numerical data of the data frame
df.describe()
```

```python
# Describe the categorical data of the data frame
df.describe(include='object')
```

```python
# Univariate: Count the frequency of model_year in sns countplot
# What are the mode_year of the cars
sns.countplot(x='model_year', data=df)
```

```python
# Bivariate: Do cars with high horsepower tend to have lower mpg?
# Use scatteplot
sns.scatterplot(x='horsepower', y='mpg', data=df)
```

```python
# Bivariate: Which origin has better mpg?
# Use boxplot for mpg
sns.boxplot(x='origin', y='mpg', data=df)
```

```python
# Bivariate: Is there a positive relationship between  mpg and displacement
# Use again scatterplot
sns.scatterplot(x='displacement', y='mpg', data=df)
```

```python
# Multivariate:  Can you check the distinct cluster of cars origin with reference to mpg and horsepower
# Use scatterplot of mpg, horsepower with hue on origin
sns.scatterplot(x='mpg', y='horsepower', hue='origin', data=df)
```

```python
# multivariate: Analyze mpg, horsepower, weight and displacement
# Fuel efficiency (mpg) goes down when horsepower, weight, or displacement goes up. That means heavier, more powerful cars usually use more fuel.
# Horsepower, weight, and displacement tend to go up together â€” powerful cars are often heavier and have larger engines.
# Use pairplot
sns.pairplot(df[['mpg', 'horsepower', 'weight', 'displacement']])
```

```python
# Multivariate; Correlation of between mpg, cylinders, displacement, horsepower, weight, accelaration, model_year
corr = df[['mpg', 'cylinders', 'displacement', 'horsepower', 'weight', 'acceleration', 'model_year']].corr()
# use heatmap
g= sns.heatmap(corr, annot=True)
g.set_title("Correlation Matrix")

# More horsepower, weight, and engine size (displacement) usually means worse fuel efficiency (lower mpg). Thatâ€™s a strong negative relationship.
# Cars with more cylinders tend to be bigger, heavier, and more powerful â€” those variables are very tightly connected.
# Acceleration improves as horsepower and weight go down. So lighter cars with smaller engines often zip around faster.
# Newer models (higher model year) tend to be more fuel-efficient, have fewer cylinders, and be lighter. That aligns with trends in eco-friendly car design.
```
