### Import and load the csv data file

```python
# import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the 'tips' dataset
df = sns.load_dataset("tips")
df
```

### 1. What’s the distribution of bills throughout the week? Reveal which day tends to have higher spending.

```python
# Show the contribution of bills throughout the week.
plt.figure(figsize=(10, 6))
sns.boxplot(x='day', y='total_bill', data=df, order=['Thur', 'Fri', 'Sat', 'Sun'])
```

### 2. How does tip amount vary by gender? Compare how tipping behavior may differ differ between male and female customers.

```python
# Display the tip amount vary by gender. Also, compare how tipping behavior may differ between male and female customers
plt.figure(figsize=(8, 6))
sns.barplot(x='sex', y='tip', data=df, errorbar=None)
plt.title('Tip Amount by Gender')
plt.xlabel('Gender')
plt.ylabel('Tip Amount')
```

### 3. Do smokers tip differently than non-smokers across time?  Reveal patterns in tipping depending on smoking status and meal time.

```python
# Display patterns in tipping depending on smoking status and meal time.
plt.figure(figsize=(10, 6))
sns.boxplot(x='time', y='tip', hue='smoker', data=df)
plt.title('Tip Amount by Time and Smoking Status')
plt.xlabel('Time of Day')
plt.ylabel('Tip Amount')
```

### 4. What’s the tipping behavior by party size and gender? Show if tipping scales with group size and whether gender plays a role in larger parties.

```python
# Show if tipping scales with group size and whether gender plays a role in larger parties using a box plot.
plt.figure(figsize=(12, 7))
sns.boxplot(x='size', y='tip', hue='sex', data=df)
plt.title('Tipping Behavior by Party Size and Gender')
plt.xlabel('Party Size')
plt.ylabel('Tip Amount ($)')
plt.show()
```

### 5. Is there a relationship between total bill and tip, broken down by day? Highlight day when generous tippers come in and how bills/tips pair.

```python
# Show the relationship between total bill and tip, broken down by day.
sns.scatterplot(data=df, x='total_bill', y='tip', hue='day')
plt.title('Relationship between Total Bill and Tip by Day')
plt.xlabel('Total Bill ($)')
plt.ylabel('Tip Amount ($)')
```

### 6. How does tipping vary between Lunch and Dinner, split by gender? Reveal how tip amounts are concentrated by time and gender. Use violin plot.

```python
# Show how tip amounts are concentrated by time and gender using a violin plot.
plt.figure(figsize=(10, 6))
sns.violinplot(x='time', y='tip', hue='sex', data=df, split=True)
plt.title('Tip Amount Distribution by Time and Gender')
plt.xlabel('Time of Day')
plt.ylabel('Tip Amount ($)')
```

### 7. Which day sees the most customers?

```python
# Find which day has the most customers by counting the occurrences of each day.
day_counts = df['day'].value_counts()
display(day_counts)

# Visualize the number of customers by day.
plt.figure(figsize=(10, 6))
sns.countplot(x='day', data=df, order=['Thur', 'Fri', 'Sat', 'Sun'])
plt.title('Number of Customers by Day')
plt.xlabel('Day of the Week')
plt.ylabel('Number of Customers')
```

### 8. What variables are most correlated with tip amount? What does it says, generally?

```python
# Calculate the correlation matrix.
correlation_matrix = df.corr(numeric_only=True)

# Display the correlation of all numerical variables with 'tip'.
display(correlation_matrix[['tip']].sort_values(by='tip', ascending=False))

# Visualize the correlation
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix of Numeric Variables')
plt.show()
```

### 9. Is party size associated with higher bills or tips?

```python
# Visualize the relationship between party size and total bill using a box plot.
plt.figure(figsize=(10, 6))
sns.boxplot(x='size', y='total_bill', data=df)
plt.title('Total Bill by Party Size')
plt.xlabel('Party Size')
plt.ylabel('Total Bill ($)')

# Visualize the relationship between party size and tip amount using a box plot.
plt.figure(figsize=(10, 6))
sns.boxplot(x='size', y='tip', data=df)
plt.title('Tip Amount by Party Size')
plt.xlabel('Party Size')
plt.ylabel('Tip Amount ($)')
```

### 10. Do total bill and tip always increase together?

```python
# Visualize the relationship between total bill and tip.
plt.figure(figsize=(8, 6))
sns.scatterplot(x='total_bill', y='tip', data=df)
plt.title('Total Bill vs. Tip Amount')
plt.xlabel('Total Bill ($)')
plt.ylabel('Tip Amount ($)')

# Calculate the correlation between total bill and tip.
correlation = df['total_bill'].corr(df['tip'])
display(f"The correlation between total bill and tip is: {correlation:.2f}")
```

### 11. How do tip and total_bill relate across days? Spot whether certain days (e.g. Saturday) have bigger bills and tips.

```python
# Show the relationship between total bill and tip across days.
plt.figure(figsize=(10, 6))
sns.scatterplot(x='total_bill', y='tip', hue='day', data=df)
plt.title('Total Bill vs. Tip Amount Across Days')
```

### 12. Are bigger groups more generous tippers? Use hue to reveal gender-based patterns among party size and tipping.

```python
# Visualize the relationship between party size and tip amount, with hue for gender.
plt.figure(figsize=(10, 6))
sns.boxplot(x='size', y='tip', hue='sex', data=df)
plt.title('Tip Amount by Party Size and Gender')
plt.xlabel('Party Size')
plt.ylabel('Tip Amount ($)')
```

### 13. Do males and females show different spending or tipping habits? Highlight behavior differences by color-coded trends.

```python
# Visualize the relationship between total bill and tip, with hue for gender.
plt.figure(figsize=(8, 6))
sns.scatterplot(x='total_bill', y='tip', hue='sex', data=df)
plt.title('Total Bill vs. Tip Amount by Gender')
plt.xlabel('Total Bill ($)')
plt.ylabel('Tip Amount ($)')
plt.show()

# Visualize the distribution of total bills by gender using a box plot.
plt.figure(figsize=(8, 6))
sns.boxplot(x='sex', y='total_bill', data=df)
plt.title('Total Bill Distribution by Gender')
plt.xlabel('Gender')
plt.ylabel('Total Bill ($)')
plt.show()

# Visualize the distribution of tips by gender using a box plot (already done in Q2, but
# including again to keep all gender-related visualizations together).
plt.figure(figsize=(8, 6))
sns.boxplot(x='sex', y='tip', data=df)
plt.title('Tip Distribution by Gender')
plt.xlabel('Gender')
plt.ylabel('Tip Amount ($)')
plt.show()
```
