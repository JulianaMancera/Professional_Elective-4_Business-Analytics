```python
# Import pandas, matplotlib, seaborn and sklearn for Linear Regression
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
```

```python
# Load the data from the CSV file into a pandas DataFrame
df = pd.read_csv("/content/SalesAdvertising.csv")
df
```

```python
# Create the feature DataFrame X and the target Series y from the df DataFrame
X = df[['advertisingSpend']]
y = df['WeeklySales']
df
```

```python
# The previous code block displayed the entire df DataFrame instead of just X and y. This code block will correctly display the first few rows of X and y.
display(X.head())
display(y.head())
```

```python
# # Create a linear regression model object and fit the model to the data
model = LinearRegression()
model.fit(X, y)
```

```python
# Show the simple linear regression equation (slope and intercept).
intercept = model.intercept_
coefficient = model.coef_[0]

print(f"Intercept: {intercept}")
print(f"Coefficient (Slope): {coefficient}")

print(f"Linear Regression Equation: y = {intercept:.2f} + {coefficient:.2f} * x")
```

```python
# Create an array of advertising spend values for prediction and use the trained model to predict weekly sales.
predict_spends = [[2000], [3000]]
predicted_sales = model.predict(predict_spends)

print(f"Predicted weekly sales for advertising spend of 2000: {predicted_sales[0]:.2f}")
print(f"Predicted weekly sales for advertising spend of 3000: {predicted_sales[1]:.2f}")
```

```python
# Plot the graph
sns.regplot(x='advertisingSpend', y='WeeklySales', data=df)
plt.xlabel('Advertising Spend')
plt.ylabel('Weekly Sales')
plt.title('Simple Linear Regression')
plt.show()
```

## Summary:
*   The model predicts weekly sales of approximately 4023.63 for an advertising spend of 2000.
*   The model predicts weekly sales of approximately 5551.24 for an advertising spend of 3000.
